Cálculo de probabilidade


```python
#Cálculo de probabilidade de obter um número par em um dado

espaco_amostral = {'1','2','3','4','5','6'}
evento_A = {'2','4','6'}
probabilidade = len(evento_A)/len(espaco_amostral)
print('Probabilidade:',probabilidade)
```

```python
#Cálculo de probabilidade de obter um número par em um dado

espaco_amostral = [1, 2, 3, 4, 5, 6]
eventos_favoráveis = [2, 4, 6]

n_E = len(eventos_favoráveis)
n_S= len(espaco_amostral)

P_E = n_E / n_S

print(f'A probabilidade de obter um número par é: {P_E:.2f} ou {P_E * 100:.0f}%')
```
